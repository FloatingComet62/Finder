```python
from ultralytics.utils import callbacks
print(callbacks.default_callbacks.keys())
```

```python
import os
import shutil
from pathlib import Path

def save_to_drive(trainer):
    epoch = trainer.epoch
    src = Path(trainer.save_dir) / 'weights' / 'last.pt'
    dst = f'/content/drive/MyDrive/sku110k-weights/epoch{epoch}.pt'
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    if src.exists():
        shutil.copy(src, dst)
        print(f"✅ Saved epoch {epoch} to Drive")
    else:
        print(f"⚠️ Could not find {src}")
```

```python
from ultralytics import YOLO

# model = YOLO("yolov8s.pt")
model = YOLO("/content/drive/MyDrive/sku110k-weights/epoch1.pt")
model.add_callback("on_train_epoch_end", save_to_drive)
results = model.train(
    data="SKU-110K.yaml",
    epochs=50,
    imgsz=640,
    device=0,
    batch=16,
    save_period=1,
)

model.save('yolo8s-sku110k.pt')
```
